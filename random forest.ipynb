```python
import math
df['Height'] = df.apply(lambda row: math.sqrt(row['Actual Weight'] / row['BMI']), axis=1)
```

```python
df['maximum Heart Rate'] = 220 - df['Age']
```

```python
import pandas as pd

# Assuming you have a DataFrame called 'data' with 'Age', 'Gender', 'Duration', 'Actual Weight', 'Height', and 'Activity Level' columns
# Replace 'data' with your actual DataFrame name

# Define a function to calculate BMR
def calculate_bmr(row):
    if row['Gender'] == 'Male':
        bmr = 88.362 + (13.397 * row['Actual Weight']) + (4.799 * row['Height'] * 100) - (5.677 * row['Age'])
    elif row['Gender'] == 'Female':
        bmr = 447.593 + (9.247 * row['Actual Weight']) + (3.098 * row['Height'] * 100) - (4.330 * row['Age'])
    else:
        bmr = 0  # Assign a default value for other gender options or handle it differently based on your requirements
    return bmr

# Define a function to calculate estimated calories burned
def calculate_calories_burned(row):
    bmr = row['BMR']  # Basal Metabolic Rate
    activity_factor = row['Exercise Intensity']  # Activity factor based on the activity or activity level
    calories_burned = bmr * activity_factor * row['Duration']
    return calories_burned

# Calculate BMR for each row
df['BMR'] = df.apply(calculate_bmr, axis=1)

# Calculate estimated calories burned for each row
df['Calories to Burn (Kcal)'] = df.apply(calculate_calories_burned, axis=1)
```
